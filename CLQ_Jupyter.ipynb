```python
import pandas as pd
```

```python
# data for each CustomerID (1 row per CustomerID)
df = pd.read_pickle('data_big.pkl')
df.head()
```

1. Which location has the maximum number of purchases been made?
- Mumbai

```python
df1 = df.groupby('CustLocation').agg(NumTrans = ('TransactionID', "count")).reset_index()
df1.sort_values(by='NumTrans', ascending=False)
```

```python
df1[df1.NumTrans == df1.NumTrans.max()]['CustLocation']
```

2. In the dataset provided, did females spend more than males, or vice versa?
- Male spend more

```python
df2 = df.groupby('CustGender').agg(TransAmnt = ('TransactionAmount (INR)', "sum")).reset_index()
df2
```

3. Report the customer with the highest average transaction amount in the dataset.

```python
df3 = df.groupby('CustomerID').agg(TransAmnt = ('TransactionAmount (INR)', "mean")).reset_index()
df3.sort_values(by='TransAmnt', ascending=False)
```

```python
df3[df3.TransAmnt == df3.TransAmnt.max()]['CustomerID']
```
